```python
yaml_content = """
train: visdrone/car/images/train
val: visdrone/car/images/val
test: visdrone/car/images/test

nc: 1  # nombre de classes
names: ['car']
"""
with open("visdrone_car.yaml", "w") as f:
    f.write(yaml_content)
```

```python

model = YOLO("yolo11n.pt")  
```

```python
model.train(
    data="visdrone_car.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="visdrone_car_detection"
) 

```
