```python
import os
from tqdm import tqdm
```

```python


# Dictionnaire des classes VisDrone
visdrone_classes = {
    0: "ignored_regions",
    1: "pedestrian",
    2: "people",
    3: "bicycle",
    4: "car",
    5: "van",
    6: "truck",
    7: "tricycle",
    8: "awning-tricycle",
    9: "bus",
    10: "motor",
    11: "others"
}

def convert_visdrone_to_yolo(visdrone_dir, output_dir, image_dir):
    os.makedirs(output_dir, exist_ok=True)
    for label_file in tqdm(os.listdir(visdrone_dir)):
        if not label_file.endswith(".txt"):
            continue
        img_name = label_file.replace(".txt", ".jpg")
        img_path = os.path.join(image_dir, img_name)

        with open(os.path.join(visdrone_dir, label_file), "r") as f:
            lines = f.readlines()

        out_path = os.path.join(output_dir, label_file)
        with open(out_path, "w") as out:
            for line in lines:
                x, y, w, h, score, cls_id, trunc, occ = map(int, line.split(","))
                if cls_id == 0:  # ignore "ignored regions"
                    continue
                x_center = (x + w / 2) / 960  # remplace 960, 540 par ta taille d’image réelle
                y_center = (y + h / 2) / 540
                out.write(f"{cls_id - 1} {x_center:.6f} {y_center:.6f} {w/960:.6f} {h/540:.6f}\n")

convert_visdrone_to_yolo(
    visdrone_dir="datasets/visdrone/annotations/train/",
    output_dir="datasets/visdrone/labels/train/",
    image_dir="datasets/visdrone/images/train/"
)
```
