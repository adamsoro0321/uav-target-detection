```python
import os

LABEL_DIR = "/media/Data/uav/uav-target-detection/dataset/visdrone/labels"
TARGET_CLASS = 4   # 'car' selon VisDrone
NEW_CLASS_ID = 0   # puisque ton dataset n’a qu’une seule classe

for split in ["train", "val"]:
    dir_path = os.path.join(LABEL_DIR, split)
    for file in os.listdir(dir_path):
        if not file.endswith(".txt"):
            continue
        path = os.path.join(dir_path, file)
        with open(path, "r") as f:
            lines = f.readlines()
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if int(parts[0]) == TARGET_CLASS:
                parts[0] = str(NEW_CLASS_ID)
                new_lines.append(" ".join(parts))
        with open(path, "w") as f:
            f.write("\n".join(new_lines))
```
